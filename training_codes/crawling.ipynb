```python
from bs4 import BeautifulSoup
import requests
import re
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
URL = "https://en.wikipedia.org/wiki/Harry_Potter_(character)"
```

```python
URL = "https://harrypotter.fandom.com/wiki/Harry_Potter"
```

```python
res = requests.get(URL)
bsr = BeautifulSoup(res.text, "html.parser")
```

```python
all_infos = bsr.find("div", {"id":"mw-content-text"}).find("div", {"class":"mw-parser-output"}).find_all("p")
```

```python
clean = re.compile('<.*?>|\n|\[(\d+)\]')
text = []
for inf in all_infos:
  result = re.sub(clean, "", str(inf))
  text.append(result)
  
  if result and "." in result and "Mr." not in result and "Ms." not in result and "Dr." not in result:
    splitted_lines = result.split(".")
    for line in splitted_lines:
      if len(line) > 5:
        text.append(line + ".")
```

```python
all_writtens=[]

for i in range(10, len(text)-1):
  if text[i-1] in text[i]:
    text[i].replace(text[i-1], "")
  if text[i+1] in text[i]:
    text[i].replace(text[i+1], "")
  all_writtens.append(text[i])

file_path="harry_info_one.json"
with open(file_path, 'w') as outfile:
  json.dump(all_writtens, outfile)
```
